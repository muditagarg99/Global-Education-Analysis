```python
#creating a function that will take a dataframe and list of countries and will return a dataframe with required columns and rows
def newdf(anydf,countrylst):
  anydf=anydf.loc[anydf['Country Name'].isin(countrylst),:]
  anydf_df_Population=anydf.loc[anydf['Topic']=='Health: Population: Dynamics',:]
  anydf_df_Population=anydf_df_Population[['Country Name', '2005', '2006','2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014']]
  anydf_df_Population.set_index(anydf_df_Population['Country Name'], inplace=True,)
  anydf_df_Population.drop('Country Name',axis=1,inplace=True)
  anydf_df_Population=anydf_df_Population.ffill(axis=1)
  anydf_df_Population=anydf_df_Population.transpose()
  return anydf_df_Population
```

## Its the time to make the list for 'Developing Countries and Developed Countries'.

```python
developing=[]
developed=[]
for index,row in df_StatsCountryData.iterrows():
  if row['Income Group'] in ['Low income','Lower middle income']:
    developing.append(row['Table Name'])
  elif row['Income Group'] in ['High income: nonOECD','Upper middle income','High income: OECD']:
    developed.append(row['Table Name'])
  else:
    pass
```

```python
#printing the list of developed and developing nations
print('Developed Nations:', developed)
print("Developing Nations:", developing)
```

```python
y = list([len(developed),len(developing)])
mylabels = ["Developed Coutries","Developing countries"]

plt.figure(figsize=(5,5))
plt.pie(y, labels = mylabels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.legend()
plt.title("percentage ratio of 'Developed Countries' and 'Developing coutries'")
plt.show() 
```

```python
#Function that will make 2 data frames one having information of developed countries and another for developing countries
def separating_nations (all_nations,ed_nations,ing_nations):
  df_ed=all_nations.copy()
  df_ing=all_nations.copy()
  for index,row in all_nations.iterrows():
    if row['Country Name'] not in ed_nations:
      df_ed.drop(index,inplace=True)
    elif row['Country Name'] not in ing_nations:
      df_ing.drop(index,inplace=True)
    else:
      df_ed.drop(index,inplace=True)
      df_ing.drop(index,inplace=True)
  return df_ed,df_ing
```

```python
# selecting the GDP data rows only and eliminating the null
data_gdp=data_gdp[data_gdp['Indicator Code']=='NY.GDP.PCAP.PP.CD']
data_gdp=data_gdp[~data_gdp['2020'].isna()]
df1,df2=separating_nations(data_gdp,developing,developed)
# df1 and df2 are for developing and developed nations respectively 
```
